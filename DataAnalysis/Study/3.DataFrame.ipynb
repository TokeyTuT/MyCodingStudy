# DataFrame 的创建

```python
from time import process_time_ns

import pandas as pd
import numpy as np
from numpy.ma.extras import column_stack

# 通过 Series 创建
a = pd.Series([1,2,3,4])
b = pd.Series([5,6,7,8])
df = pd.DataFrame({"第一列":a,"第二列":b})
print(df["第一列"])
# 通过字典来创建
df2 = pd.DataFrame(
    {
        'name':['jak','tokey','mike','joe','amy'],
        'age':[10,20,30,40,50]
     },index=['0001','0002','0003','0004','0005'],
    columns=['name','age']
)
df2
```

# DataFrame 的访问

```python
df = df2
# 获取单列数据
print(df.name) # Series
print(df['name']) # Series

print(df[['name']])  # DataFrame类型
```

```python
# 获取多列数据
df[['name','age']] # DataFrame
```

```python
# 查看部分数据
print(df.head(2))
print(df.tail(2))
```

```python
# 使用布尔索引筛选
print(df[df.age < 50])
```

```python
data = pd.DataFrame(
{
    '姓名':['张三','李四','王五','赵六','钱七'],
    '数学':[85,92,78,88,95],
    '英语':[90,88,85,92,80],
    '物理':[75,80,88,85,90]
})
data['总分'] = data[['数学','英语','物理']].sum(axis="columns")
data['平均分'] = data['总分'] / 3
d2 = data[(data['数学'] > 90) | (data['英语']>85)]
print(d2)
d3 = data.sort_values('总分',ascending=False)
print(d3)
```

```python
data = pd.DataFrame(
    {
        '产品名称':['A','B','C','D'],
        '单价':[100,150,200,120],
        '销量':[50,30,20,40]
    }
)
data['总销售额'] = data['单价']*data['销量']
data.nlargest(1,columns=['总销售额'])
data.sort_values('总销售额',ascending=False)
```

```python
data = pd.DataFrame(
    {
        '用户 ID':[101,102,103,104,105],
        '用户名':['Alice','Bob','Charlie','David','Eve'],
        '商品类别':['A','B','C','D','E'],
        '商品单价':[1200,300,800,150,200],
        '购买数量':[1,3,2,5,4]
    }
)
data['总销售额'] = data['商品单价']*data['购买数量']
data.nlargest(1,'总销售额')
print(data['总销售额'].mean())
print(data['购买数量'].sum())
```
