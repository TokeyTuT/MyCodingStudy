```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['STHeiti']
penguins = pd.read_csv('data/penguins.csv')
penguins.info()
```

```python
# 直方图
sns.histplot(data=penguins,x="species")
```

```python
# 核密度估计图
'''
核密度估计图（Kernel Density Estimation, KDE）
是一种用于估计连续随机变量概率密度函数的无参数方法
简单来说，它就像是平滑版的直方图。
'''
sns.kdeplot(data=penguins,x="bill_length_mm")
```

```python
# 在直方图上绘制核密度估计图
sns.histplot(data=penguins,x="bill_length_mm",kde=True)
```

```python
# 计数图
sns.countplot(data=penguins,x="island")
```

```python
sns.histplot(data=penguins,x='island')
```

```python
# 散点图
sns.scatterplot(data=penguins,x='body_mass_g',y='bill_length_mm',hue='sex') # hue 表示分组的聚合
```

```python
# 蜂窝图
sns.jointplot(data=penguins,x='body_mass_g',y='bill_length_mm',kind="hex") # 通过 jointplot()函数，设置 kind='hex'绘制蜂窝图
```

```python
# 成对关系图
'''
成对关系图是一种用于显示多个变量之间关系的可视化工具。
它可以展示各个变量之间的成对关系，并且通过不同图标形式帮助我们理解数据中各个变量之间的相互作用。
对角线上的图通常显示每个变量之间的分布(如直方图或核密度估计图)，帮助我们观察每个变量之间特性。
其他位置展示所有变量之间的两两关系，用散点图表示。
'''
sns.pairplot(data=penguins,hue='species')
```
