# 折线图

```python
import random

# 绘制折线图
import matplotlib.pyplot as plt
# 配置中文库，防止报错
from matplotlib import rcParams # 控制字体
from matplotlib.lines import lineStyles
from pandas.io.sas.sas_constants import page_comp_type
from pyparsing import alphas
from wcwidth import width

rcParams['font.sans-serif'] = 'STHeiti' # mac:STHeiti win:SimHei
# 创建一个图标，设置大小
plt.figure(figsize=(10, 5)) # (width,height)
# 设计折线图
month_lst = ['一月','二月','三月','四月']
sales = [100,200,150,250]
plt.plot(month_lst,sales,
         label='产品 A',
         linewidth=2,
         linestyle='--',
         color='orange',
         marker='o'
         ) # label 图例 marker 控制是否为原点 linewidth 线的粗细
plt.title('2026年销售趋势',color='r',fontsize=20)
#设置坐标轴标签
plt.xlabel('月份',fontsize=10)
plt.ylabel('销售额(万元)',fontsize=10)
# 添加图例
plt.legend(loc='upper left')
# 添加网格线
# plt.grid(True)
plt.grid(axis='y',alpha=0.1,color='black',linestyle='--') # alpha 透明度 linestyle 网格线的样式 ‘-’->实线，‘--’->虚线
# 设置刻度字体大小
plt.xticks(rotation=0,fontsize=12) # rotation 当数据比较密的时候，可以让刻度字体旋转，这个参数代表逆时针往上旋转的角度
# 设置 y 轴的范围
plt.ylim(0,300)
# 在每个数据点上显示数值
for x,y in zip(month_lst,sales):
    plt.text(x,y + 5,str(y),ha='center',va='bottom',fontsize=10)
# 显示图表
plt.show()
```

# 柱形图

```python
# 条形图 bar
subject = ['数学','语文','英语','物理']
score = [80,85,90,95]
plt.bar(subject,score,
        label='Tokey',
        width=0.5,
        color='orange'
        )
plt.title('Tokey的成绩分布',fontsize=20,color='red')
plt.xlabel('科目')
plt.ylabel('成绩')
plt.grid(axis='y',alpha=0.1)
plt.ylim(0,100)
for x,y in zip(subject,score):
    plt.text(x,y + 1,str(y),ha='center',va='bottom',fontsize=10)
plt.legend()
# 自动优化排版
plt.tight_layout()
plt.show()
```

# 条形图

```python
# 数据
subject = ['数学','语文','英语','物理']
score = [80, 85, 90, 95]

plt.figure(figsize=(10, 5))

# 绘制水平条形图
plt.barh(subject, score, label='Tokey', color='orange')

# 标题与标签 (注意 X/Y 轴已经互换)
plt.title('Tokey的成绩分布', fontsize=20, color='red')
plt.xlabel('成绩')
plt.ylabel('科目')

# 网格线：水平图中通常对 X 轴（垂直线）做网格
plt.grid(axis='x', alpha=0.2)

# 设置 X 轴范围（成绩范围）
plt.xlim(0, 110)

# 核心修正：遍历数据添加数值标签
for s, v in zip(subject, score):
    # 参数含义：(x轴位置, y轴位置, 内容)
    # v + 1 表示在条形末端往右偏移 1 个单位
    plt.text(v + 1, s, str(v), ha='left', va='center', fontsize=12)

plt.legend()
plt.tight_layout()
plt.show()
```

# 饼图

```python
# 饼图
things=['学习','娱乐','运动','睡觉','其他']
times=[6,3,1,8,6]
plt.pie(times,labels=things,
        autopct='%.1f%%', # 显示百分比
        startangle=30, # 调整初始画图的角度
        wedgeprops={'width':0.66}, # 设置中间镂空的圆环
        pctdistance=0.65 # 设置百分比的位置
        )
plt.title('每天的时间分布',color='red',fontsize=20)
plt.text(0,0,'总计\n100%',ha='center',va='bottom',fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# 爆炸式的饼图
times=[6,3,1,8,6]
plt.pie(times,labels=things,
        autopct='%.1f%%', # 显示百分比
        startangle=30, # 调整初始画图的角度
        pctdistance=0.65, # 设置百分比的位置
        explode=[0.11,0,0,0,0.1] # 设置突出块
        ,shadow=True # 设置阴影
        )
plt.title('每天的时间分布',color='red',fontsize=20)
plt.text(0,0,'总计\n100%',ha='center',va='bottom',fontsize=10)
plt.tight_layout()
plt.show()
```

# 散点图

```python
import random

plt.figure(figsize=(10,8))
# 散点图
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0,10)
    x.append(tmp)
    tmp2=2*tmp + random.gauss(0,2)
    y.append(tmp2)
plt.scatter(x,y,
            color='orange',
            alpha=0.5,
            s=20 # 原点的大小
            )
plt.xlabel('x',fontsize=20)
plt.ylabel('y 的值',fontsize=20)
plt.grid(axis='y',alpha=0.1)
# 画出回归线
plt.plot([0,10],[0,20],color='r',linewidth=2,linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# 案例
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
rcParams['font.sans-serif'] = 'STHeiti'

df = pd.read_csv('data/weather.csv',parse_dates=['date'])
# df.head()
# df.info()
df = df[df['date'].dt.year == 2015]
df['temp_arg'] = (df['temp_max'] + df['temp_min'])/2

# 绘制一个气温的变化图
plt.figure(figsize=(10,10))
plt.plot(df['date'],df['temp_max'],label='最高气温')
plt.plot(df['date'],df['temp_min'],label='最低气温')
plt.plot(df['date'],df['temp_arg'],label='平均气温')
plt.legend()
plt.title('2015年每月最高/最低气温变化',color='orange',fontsize=20)
plt.ylabel('气温(℃)',fontsize=20)
plt.xlabel('日期(年-月)',fontsize=20)
plt.tight_layout()
```
