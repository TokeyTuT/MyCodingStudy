# Series

```python
import pandas as pd
s = pd.Series([10,1,2,3,4,5,6])
print(s)  # 索引-值

# 自定义索引值
s = pd.Series([10,1,2,3,4],index=['A','B','C','D','E'])
print(s)
# 定义标签
s = pd.Series([10,1,2,3,4],index=['A','B','C','D','E'],name='Tokey')
print(s)
```

```python
# 通过字典创建
s = pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})
print(s)

s1 = pd.Series(s,index=['a','b','c','d','e'],dtype='int')
print(s1)

s2 = pd.Series(s1,index=['a','b'])
print(s2)
```

# Series 的属性

```python
arr = pd.Series([1,2,3],index=['a','b','c'],name='test')
print(arr)
# index，获取索引对象
print(arr.index)
# value，获取值
print(arr.values)
# shape，获取形状
print(arr.shape,arr.ndim,arr.size)
print(arr.dtypes,arr.name)
```

```python
# 显式索引 —— 支持切片
print(arr.loc['a':'c'])

print(arr.at['a']) # at 不支持切片
# 隐式索引 —— 支持切片
print(arr.iloc[0:2])

print(arr.iat[0]) # at 不支持切片
```

## Series 的访问

```python
# print(arr[1]) 报错，要用标签访问
print(arr['a'])
# 支持布尔索引
print(arr[arr < 2])
# head() —— 默认返回数据的前五行
print(arr.head())
print(arr.head(2)) # 返回数据的前 2 行
# tail() —— 默认返回数据的末五行
print(arr.tail())
print(arr.tail(1)) # 返回数据的最后一行
```
